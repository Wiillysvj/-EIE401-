Debes generar vectores que representan intervalos de tiempo o de una variable.

$$

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
# Importamos librerías necesarias
import numpy as np
import matplotlib.pyplot as plt


# Función 1: f(x) = e^(-x^2)

# Vector de x que recorre intervalo [-5, 5]
x = np.linspace(-5, 5, 400)

# Evaluamos la función en cada punto del vector
fx = np.exp(-x**2)

# Creamos la gráfica de f(x)
plt.figure(figsize=(8,6))
plt.plot(x, fx, label=r'$f(x)=e^{-x^2}$')
plt.title('Gráfica de $f(x)=e^{-x^2}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()


# Función 2: f(z) = y*sin(x)

# Se asume que "y" es una constante.
# Por ello, podemos definir y = 2 y que x varía en [0, 2π]

x2 = np.linspace(0, 2*np.pi, 400)
y = 2
fz = y * np.sin(x2)

# Creamos la gráfica de f(z)
plt.figure(figsize=(8,6))
plt.plot(x2, fz, label=r'$f(z)=y\sin(x)$ con $y=2$')
plt.title('Gráfica de $f(z)=y\sin(x)$')
plt.xlabel('x')
plt.ylabel('f(z)')
plt.grid(True)
plt.legend()
plt.show()
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

R: El ruido rosa es un tipo de señal aleatoria en la que la potencia disminuye inversamente a la frecuencia (1/f), lo que hace que cada banda de octava tenga la misma energía. Esto resulta en un sonido más equilibrado y natural que el ruido blanco.

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para generar ruido rosa a partir de ruido blanco
def generate_pink_noise(N):
    white = np.random.randn(N)
    X = np.fft.rfft(white)
    # Escalamos las frecuencias (evitamos el DC)
    k = np.arange(1, len(X))
    X[1:] /= np.sqrt(k)
    pink = np.fft.irfft(X, n=N)
    return pink / np.std(pink)

N = 2048  # Número de muestras

# Generamos los ruidos
white_noise = np.random.randn(N)
pink_noise = generate_pink_noise(N)

# Gráfica 1: Dominio del Tiempo
plt.figure(figsize=(10, 4))
plt.plot(white_noise, label='Ruido Blanco')
plt.plot(pink_noise, label='Ruido Rosa')
plt.title('Ruido en el Dominio del Tiempo')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
